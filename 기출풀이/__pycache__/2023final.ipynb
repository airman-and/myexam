```python
class R:
    def __init__(self, width, height):
        self.width = width
        self.height = height
    
    def area(self):
        return self.width*self.height
    
    def __str__(self):
        return f'({self.width},{self.height})'
    
    def __eq__(self,other):
        return self.area() == other.area()
    
```

```python
def myEnumerate(iterable, start):
    for i in range(len(iterable)):

        if (i % 2) == (start % 2):
            yield (i, iterable[i])

print(list(myEnumerate("abcdef", 0)))
print(list(myEnumerate([2, 3, 4, 5, 6], 3)))
print(list(myEnumerate(['Spring','Summer', 'Fall', 'Winter'], 1)))
```

```python
def fun(n):
    List = list(range(1,n+1,2))
    Sum = sum(List)
    if n >= 11:
        return f'{List[0]}+{List[1]}+{List[2]}+...+{List[-2]}+{List[-1]}={Sum}'
    else:
        SList = list(map(str, List))
        S = '+'.join(SList)
        return f'{S}={Sum}'

fun(12)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

data = {
    'pets': [np.nan, 1, 3, 2, 3],
    'children': [2,3,np.nan,4,3]
}

index  = ['Kim', 'Lee', 'Park', 'Choi', 'Jang']

df = pd.DataFrame(data, index=index)
df.index.name = 'name'
df.plot(kind='bar')

plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0,40,500)
y1 = x
y2 = x**2/40
y3 = x**3/900

plt.plot(x,y1,label = 'x')
plt.plot(x,y2,label = 'x^2/40')
plt.plot(x,y3,label = 'x^3/900')

plt.title('Functions')
plt.xlabel('x axis')
plt.legend(loc = 'lower right')

plt.show()
```

```python
import tkinter as tk

window = tk.Tk()
window.title("")
window.resizable(False, False)

counter = tk.IntVar(value=0)

def change_value(amount):
    try:
        counter.set(counter.get() + amount)
    except tk.TclError:
        counter.set(amount)

entry = tk.Entry(window, width=22, justify="right", textvariable=counter)
entry.grid(row=0, column=0, columnspan=2, padx=5, pady=5)

btn_up = tk.Button(window, text="카운터업", command=lambda: change_value(1), width=10)
btn_up.grid(row=1, column=0, padx=5, pady=5)

btn_down = tk.Button(window, text="카운터다운", command=lambda: change_value(-1), width=10)
btn_down.grid(row=1, column=1, padx=5, pady=5)

btn_reset = tk.Button(window, text="리셋", command=lambda: counter.set(0), width=22)
btn_reset.grid(row=2, column=0, columnspan=2, padx=5, pady=5)

window.mainloop()
```
